 <p>Now we will implement the <font color='green'> Tylor Series </font> for given function $f(x)$.

 

```
def tylor(x1,n):
  sum=0
  for i in range(n+1):
    s=sympy.Symbol('x')
    y=sympy.lambdify(s,series[i],'numpy')
    t=y(x1)
    sum=sum+(t*pow(.1,i))/math.factorial(i)
  return sum
```

Now calculating the approximate errors.

```
def aprroximate_error(n):

  xi=tylor(3,n-1)
  xf=tylor(3,n)

  return (abs((xf-xi)/xf))*100
```

```
axis_x = [2,3,4,5,6,7,8]
axis_y=[]
for i in axis_x:
  axis_y.append(aprroximate_error(i))

plt.plot(axis_x,axis_y,color='#8E44AD',marker='.',markersize=7)
plt.xlabel("n")
plt.ylabel("Relative Approximate Error")
```

<font color="blue"> Result:</font> We can see that, increasing of n takes the error more close to 0.
