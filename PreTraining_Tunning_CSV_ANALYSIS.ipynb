```python

import pandas as pd
import os
import glob


csv_files = glob.glob(os.path.join("/content/non_random", "*.csv"))

dic={}
i=0

# loop over the list of csv files
for f in csv_files:
	k = f.split("/")
	print("File Name: ")
	print(k[-1])
	# read the csv file
	df = pd.read_csv(f)
	dic[i] = df
	# print the content
	print('Content:')
	display(df)
	print("-----------------------------------------------------------------------------------------------------------------------------")
	print()
	i+=1
```
