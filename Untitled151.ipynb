```python
caseids = list(
    set(df_trks[df_trks['tname'] == 'Solar8000/ST_II']['caseid']) &
    set(df_trks[df_trks['tname'] == 'Solar8000/ART_MBP']['caseid']) &
     set(df_trks[df_trks['tname'] == 'Solar8000/ART_SBP']['caseid'])&
     set(df_trks[df_trks['tname'] == 'Solar8000/ART_DBP']['caseid'])&
    set(df_trks[df_trks['tname'] == 'SNUADC/ECG_II']['caseid'])
)

print('Total {} cases found'.format(len(caseids)))
```

```python
vals = vitaldb.load_case(caseids[0], ['SNUADC/ECG_II','Solar8000/ST_II','Solar8000/ART_MBP','Solar8000/ART_SBP','Solar8000/ART_DBP'], 1 / 100)
```

```python
time_col=[]
for i in range(len(vals)):
  time_col.append(i*(1/500))

vals = np.insert(vals, 0, time_col, axis=1)
vals = np.insert(vals, 0, time_col, axis=1)
vals = np.insert(vals, 0, time_col, axis=1)

df = pd.DataFrame(vals, columns=['Segment','Time(sec)','Time','SNUADC/ECG_II','Solar8000/ST_II','Solar8000/ART_MBP','Solar8000/ART_SBP','Solar8000/ART_DBP'])
```

```python
indx = 0

for i in tqdm(df['Time(sec)']):
  df['Time'][indx] = str(timedelta(seconds=i))
  indx+=1
```

```python

sec_segment = 10
segment_num = 1
start_idx = 0
end_idx = int(sec_segment/(1/500) + 1)





for _ in tqdm(range(int(df.shape[0]/int(sec_segment/(1/500) + 1)))):

  temp_df = df.iloc[start_idx:end_idx,:]

  mean_1 = temp_df["Solar8000/ST_II"].mean()
  mean_2 = temp_df["Solar8000/ART_MBP"].mean()
  mean_3 = temp_df["Solar8000/ART_SBP"].mean()
  mean_4 = temp_df["Solar8000/ART_DBP"].mean()


  for i in tqdm(range(start_idx,end_idx)):
    df["Segment"][i] = "Segment_"+str(segment_num)
    df["Solar8000/ST_II"][i] = mean_1
    df["Solar8000/ART_MBP"][i]= mean_2
    df["Solar8000/ART_SBP"][i] = mean_3
    df["Solar8000/ART_DBP"][i] = mean_4


  start_idx = end_idx
  end_idx = end_idx + int(sec_segment/(1/500))
  segment_num += 1
```
