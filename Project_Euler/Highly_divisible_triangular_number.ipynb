```
def factors(n):
  list=[]
  
  for i in range(1,int(n/2)+1):
    if n%i==0:
      list.append(i)
  list.append(n)
  return len(list)
```

```
print(factors(1000))
```

```
import timeit
start=timeit.default_timer()
s=1
p=0
while True:
  p=p+s
  print(p,factors(p))
  if factors(p)>500:
    print(p)
    break
  s+=1
stop=timeit.default_timer()

print("Time: ",(stop-start))
```
