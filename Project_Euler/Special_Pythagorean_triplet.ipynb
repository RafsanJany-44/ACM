```
m=int(input())
n=int(m/2)
k=0
l=0
for i in range(1,int(n/2)):
  if n%i==0:
      if i>(n/i-i):
        k=i
        l=int(n/i-i)
        break

a=pow(k,2)-pow(l,2)
b=2*k*l
c=pow(k,2)+pow(l,2)
p=a+b+c
if pow(a,2)+pow(b,2)==pow(c,2) and p==m:
  print(m)
  print("a+b+c: ",p)
  print("abc: ",(a*b*c))
  print(a,b,c)
else:
  print("No Numbers exists!")

```
