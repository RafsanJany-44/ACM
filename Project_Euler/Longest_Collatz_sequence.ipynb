```
def p(n):
  li=[]
  while True:
    if n%2==0:
      n=n/2
    else:
      n=3*n+1
    li.append(n)
    if n==1:
      break
  return (len(li))
```

```
n=1000000
li2=[]
li3=[]
for i in range(1000000):
  o=p(n)
  li2.append(n)
  li3.append(o)
  print(n,o)
  n=n-1
print(li2[li3.index(max(li3))])
print(li3)
```
