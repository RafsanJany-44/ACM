```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
%matplotlib inline
```

```python
import random

data_table={
    "x": random.sample(range(1, 150), 40),
    "y": random.sample(range(1, 150), 40)
}

data=pd.DataFrame(data_table)



plt.scatter(
   data["x"], data['y'],
   c='white', marker='o',
   edgecolor='red', s=30
)
plt.show()
```

```python
data.head()
```

```python
num_of_cluster = 6
kmeans = KMeans(n_clusters=num_of_cluster)
kmeans.fit(data)

centroids = kmeans.cluster_centers_
labels = kmeans.labels_

print(centroids)
print(labels)
```

```python
data['labels'] = labels
data
```

```python
data['labels'] = labels




import random
get_colors = lambda n: ["#%06x" % random.randint(0, 0xFFFFFF) for _ in range(n)]
get_colors(num_of_cluster)





initial_group = data[data['labels']==0].plot( kind='scatter', x='x', y='y', color='DarkGreen', label="Cluster 1" )

for i in range(1,num_of_cluster):
   data[data['labels']==i].plot( kind='scatter', x='x', y='y', color=get_colors(num_of_cluster)[i], ax=initial_group, label="Cluster "+str(i+1) )







initial_group.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05),
          ncol=3, fancybox=True, shadow=True)


plt.scatter(centroids[:, 0],centroids[:, 1], marker = "*",s=150, linewidths = 2, zorder = 10)

plt.show()
```
