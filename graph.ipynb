```python
import pandas as pd


data = pd.read_csv("/content/training_log.csv")


data.head()

x = list(data["epoch"])
y1 = list(data["accuracy"])
y2 = list(data['loss'])
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize = (10,7))


plt.plot(x, y1, '1-', label='Points removed')
plt.plot(x, y2, 'D-', label='Masked values')
#plt.plot(x*1.0, y4, '1-', label='NaN values')

plt.xlabel('Epoch')
plt.ylabel('Accueacy')
plt.title('set tittle')
plt.legend(loc="lower right")
plt.show()
     
```
