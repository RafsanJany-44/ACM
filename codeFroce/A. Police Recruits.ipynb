```
t=int(input())
talika=list(map(int,input().split()))
flag=0
count=0
u=0
for i in range(t):
  if talika[i]== -1 and flag==0:
    count+=1
  elif talika[i]== -1 and flag==1:
    u=u+talika[i]
    if u==0:
      flag=0
    else:
      flag=1
  elif talika[i]!=-1:
    u=u+talika[i]
    flag=1
print(count)
```
