```
def dis(n):
  if int((n**.5) + 0.5) ** 2==n:
    if moulik(int(n**.5))==1:
      return "YES"
    else:
      return "NO"
  else:
    return "NO"
```

```
def moulik(n):
  flag=1
  for i in range(2,int(n/2)):
    if n%i==0:
      flag=0
      break
  if n==4:
    flag=0
  return flag
```

```
print(moulik(2))
```

```
n=int(input())
talika=list(map(int,input().split()))
talika2=[]
for i in range(n):
  if talika[i]==1:
    talika2.append("NO")
  else:
    talika2.append(dis(talika[i]))
print(*talika2, sep='\n') 
```

```
def m(n):
  list1=[1]
  for i in range(1,n):
    if n%i==0:
        list1.append((n/i))
  return list1
```
