```
#               15
#               77 68 96 60 92 75 61 60 66 79 80 65 60 95 92
#                  4
#7    10 10 58 31 63 40 76 
n=int(input())
talika=list((map(int,input().split())))
boroTa=max(talika)
chotoTa=min(talika)
potaka=0
potaka2=0
boro_shuchi=n+1
print(talika)
if len(talika)==2:
  if talika[0]>talika[1]:
    print(0)
  else:
    print(1)
elif len(talika)==1 or chotoTa==boroTa:
  print(0)
else:
  for i in range(n):
    if talika[i]==boroTa and potaka==0:
      boro_shuchi=i
      potaka=1
    elif talika[i]==chotoTa:
      choto_shuchi=i
      if i==boro_shuchi:
        potaka2=1
  print(boro_shuchi,choto_shuchi)
  if potaka2==1:
    print(0)
    print(boro_shuchi+n-choto_shuchi-3)
  else:
    print(1)
    print(boro_shuchi+n-choto_shuchi-1)
    print(talika[-1::-1])
 
```
