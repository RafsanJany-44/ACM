```
t=int(input())
for _ in range(t):
  f(int(input()))
def f(k):
  if k%10!=0:
    talika=str(k)
    g=len(talika)-1
    talika2=[]
    for i in range(g+1):
      u=(int(k/(10**(g-i))))
      w=u*(10**(g-i))
      talika2.append(w)
      k=k%w
    print(len(talika2))
    print(*talika2,sep=' ')
  else:
    print(1)
    print(k)
```
