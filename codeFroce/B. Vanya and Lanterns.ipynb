```
n,l=map(int,input().split())
list1=sorted(set(list(map(int,input().split()))))
list2=[]

if len(list1)==1:
  if list1[0]<l-list1[0]:
    print(l-list1[0])
  else:
    print(list1[0])

else:
  a=list1[0]
  b=l-list1[-1]

  for i in range(len(list1)-1):
    list2.append((list1[i+1]-list1[i])/2)
  m=max(list2)
  if 0 in list1 and l in list1:
    print("%.10f" % m)
  elif l in list1:
    print("%.10f" % m) if m>a else print(a)
  elif 0 in list1:
    print("%.10f" % m) if m>b else print(b)
  else:
    if m>=a and m>=b:
      print("%.10f" % m)
    elif a>=m and a>=b:
      print("%.10f" % a)
    elif b>=m and b>=a:
      print("%.10f" % b)
```
