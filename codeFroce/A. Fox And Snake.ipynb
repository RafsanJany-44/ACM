```
n,m=map(int,input().split())
p=""
t=""
s=""
k=1
for i in range(m):
  p=p+"#"
  if i<m-1:
    t=t+"."
  else:
    t=t+'#'
  if i==0:
    s=s+"#"
  else:
    s=s+"."
for i in range(1,n+1):
  if i%2!=0:
    print(p)
  else:
    if k==1:
      print(t)
      k=2
    else:
      print(s)
      k=1
```
