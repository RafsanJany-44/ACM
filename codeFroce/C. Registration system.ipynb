```
n=int(input())
talika=[]

for i in range(n): talika.append(input())

def count(talika,r):
  talika2=[]
  p=1
  if talika[0][-1].isnumeric() is True:
    return talika
  else:
    for i in range(r+1):
      talika2.append(talika[i])
    for i in range(r,n-1):
      if talika[r]==talika[i+1]:
        talika[i+1]=talika[i+1]+str(p)
        talika2.append(talika[i+1])
        p+=1
      else:
        talika2.append(talika[i+1])
    return talika2

y=talika
for i in range(n):
  y=count(y,i)

for i in range(n):
  if y[i][-1].isnumeric() is False:
    y[i]="OK"
print(*y,sep='\n')
```

```
ovidhan = {}
talika=[]
for _ in range(int(input())):  # we will not use that variable
    nam = input()
    if nam in ovidhan:
        ovidhan[nam] += 1
        talika.append(nam + str(ovidhan[nam]))
    else:
        ovidhan[nam] = 0
        talika.append("OK")
print(*talika,sep="\n")
```
