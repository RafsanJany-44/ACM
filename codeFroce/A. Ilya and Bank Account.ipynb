```
n=int(input())
if n>=0:
  print(n)
else:
  if n>-10:
    if len(str(n))==3:
      print('0')
    else:
      print(n)
  else:
    s=str(n).replace('-','')
    l=len(s)
    r=int(s[:l-1])
    t=int(s[:l-2]+s[l-1])
    if r==0 or t==0:
      print(0)
    elif r<t:
      print('-'+str(r))  
    else: 
      print('-'+str(t))
```
