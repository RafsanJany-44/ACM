```
talika=[]
for i in range(1,int(input())+1):talika.append(i)
x=list(map(int,input().split()))
y=list(map(int,input().split()))
talika2=list(set(x+y))
if talika2==talika:print("I become the guy.")
elif talika2!=talika or x[0]==0 or y[0]=='0':print("Oh, my keyboard!")
```
