```
def new(n):
  if n%4!=0:
    print("NO")
  else:
    talika1=[]
    talika2=[]
    i=1
    j=2
    m=n/2
    r=1+2*(m-1)
    last=int(r+(m/2)*(2*2+2*(m-1))-(m/2)*(2*1+2*(m-1)))
    for _ in range(int(n/2)):
      talika1.append(j)
      if i!=r:
        talika2.append(i)
      else:
        talika2.append(last)
      i+=2
      j+=2
    print("YES")
    print(*(talika1+talika2),sep=' ')
```
